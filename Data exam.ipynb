```python

import polars as pl

# Load all three files
event_df   = pl.scan_parquet("event_full_2025.parquet")
trans_df   = pl.scan_parquet("transaction_full_2025.parquet")
items_df   = pl.scan_parquet("items.parquet")

# ---- Schema ----
print("=== event_full_2025 schema ===")
print(event_df.schema)

print("\n=== transaction_full_2025 schema ===")
print(trans_df.schema)

print("\n=== items schema ===")
print(items_df.schema)
```

```python

# ---- Statistics: row counts, date ranges ----
# Event
e_stats = event_df.select([
    pl.len().alias("rows"),
    pl.col("event_date").min().alias("date_min"),
    pl.col("event_date").max().alias("date_max"),
    pl.col("customer_id").n_unique().alias("unique_customers"),
    pl.col("item_id").n_unique().alias("unique_items"),
    pl.col("event_type").n_unique().alias("unique_event_types"),
]).collect()
print("=== event_full_2025 stats ===")
print(e_stats)

# Event types
e_types = event_df.select(pl.col("event_type")).collect()["event_type"].value_counts().sort("count", descending=True)
print("\nEvent types distribution:")
print(e_types)

# Transaction
t_stats = trans_df.select([
    pl.len().alias("rows"),
    pl.col("updated_date").min().alias("date_min"),
    pl.col("updated_date").max().alias("date_max"),
    pl.col("customer_id").n_unique().alias("unique_customers"),
    pl.col("item_id").n_unique().alias("unique_items"),
]).collect()
print("\n=== transaction_full_2025 stats ===")
print(t_stats)

# Items
i_stats = items_df.select([
    pl.len().alias("rows"),
    pl.col("item_id").n_unique().alias("unique_items"),
]).collect()
print("\n=== items stats ===")
print(i_stats)
print("\nCategory L1 distribution:")
print(items_df.select(pl.col("category_l1")).collect()["category_l1"].value_counts().sort("count", descending=True))
```

```python

# Compact summary
e_s = event_df.select([
    pl.len().alias("rows"),
    pl.col("event_date").min().alias("date_min"),
    pl.col("event_date").max().alias("date_max"),
    pl.col("customer_id").n_unique().alias("n_customers"),
    pl.col("item_id").n_unique().alias("n_items"),
]).collect()
t_s = trans_df.select([
    pl.len().alias("rows"),
    pl.col("updated_date").min().alias("date_min"),
    pl.col("updated_date").max().alias("date_max"),
    pl.col("customer_id").n_unique().alias("n_customers"),
    pl.col("item_id").n_unique().alias("n_items"),
]).collect()
i_s = items_df.select([pl.len().alias("rows"), pl.col("item_id").n_unique().alias("n_items")]).collect()

print("EVENT:")
for col in e_s.columns:
    print(f"  {col}: {e_s[col][0]}")
print("\nTRANSACTION:")
for col in t_s.columns:
    print(f"  {col}: {t_s[col][0]}")
print("\nITEMS:")
for col in i_s.columns:
    print(f"  {col}: {i_s[col][0]}")

# Monthly distribution in transactions
monthly = trans_df.select([
    pl.col("updated_date").dt.month().alias("month"),
]).collect()["month"].value_counts().sort("month")
print("\nTransaction monthly distribution:")
print(monthly)
```

```python

# ================================================================
# DATA SPLIT:  Train = tháng 1-10 | Val = tháng 11 | Test = tháng 12
# Cột thời gian: transaction -> updated_date | event -> event_date
# ================================================================

# --- Transaction split ---
trans_train = trans_df.filter(pl.col("updated_date").dt.month().is_between(1, 10))
trans_val   = trans_df.filter(pl.col("updated_date").dt.month() == 11)
trans_test  = trans_df.filter(pl.col("updated_date").dt.month() == 12)

# --- Event split (dùng event_date, chỉ lấy năm 2025) ---
event_2025 = event_df.filter(pl.col("event_date").dt.year() == 2025)
event_train = event_2025.filter(pl.col("event_date").dt.month().is_between(1, 10))
event_val   = event_2025.filter(pl.col("event_date").dt.month() == 11)
event_test  = event_2025.filter(pl.col("event_date").dt.month() == 12)

# --- In tóm tắt ---
def split_summary(name, train_lf, val_lf, test_lf):
    tr = train_lf.select(pl.len()).collect()[0, 0]
    v  = val_lf.select(pl.len()).collect()[0, 0]
    te = test_lf.select(pl.len()).collect()[0, 0]
    total = tr + v + te
    print(f"\n{'='*45}")
    print(f"  {name}")
    print(f"{'='*45}")
    print(f"  Train  (tháng 1-10) : {tr:>10,} rows  ({tr/total*100:.1f}%)")
    print(f"  Val    (tháng 11)   : {v:>10,} rows  ({v/total*100:.1f}%)")
    print(f"  Test   (tháng 12)   : {te:>10,} rows  ({te/total*100:.1f}%)")
    print(f"  Total               : {total:>10,} rows")

split_summary("TRANSACTION", trans_train, trans_val, trans_test)
split_summary("EVENT (2025 only)", event_train, event_val, event_test)

# Unique customers per split (transaction)
tr_custs = trans_train.select(pl.col("customer_id").n_unique()).collect()[0, 0]
v_custs  = trans_val.select(pl.col("customer_id").n_unique()).collect()[0, 0]
te_custs = trans_test.select(pl.col("customer_id").n_unique()).collect()[0, 0]
print(f"\nTransaction unique customers:")
print(f"  Train: {tr_custs:,} | Val: {v_custs:,} | Test: {te_custs:,}")
```

```python
import json
from pathlib import Path

# Collect tháng 12 (nếu vẫn là LazyFrame)
test_df = trans_test.collect() if hasattr(trans_test, "collect") else trans_test

# Build ground truth dict: {customer_id: [item_id, ...]}  (unique items per customer)
gt_rows = (
    test_df
    .group_by("customer_id")
    .agg(pl.col("item_id").unique().alias("items"))
)
ground_truth: dict[str, list] = {
    str(row["customer_id"]): sorted(row["items"])
    for row in gt_rows.iter_rows(named=True)
}

# Save
out_dir = Path("outputs/predictions")
out_dir.mkdir(parents=True, exist_ok=True)
out_path = out_dir / "ground_truth_month12.json"
with open(out_path, "w") as f:
    json.dump(ground_truth, f)

print(f"Ground truth saved → {out_path}")
print(f"Total customers : {len(ground_truth):,}")
print(f"Sample (3 entries):")
for cid, items in list(ground_truth.items())[:3]:
    print(f"  customer {cid}: {items[:5]}{'…' if len(items) > 5 else ''}")
```
